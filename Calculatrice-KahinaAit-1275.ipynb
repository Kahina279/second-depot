```python
import sys
import math
from PyQt5 import QtCore, QtWidgets, QtGui
from PyQt5.QtWidgets import QMainWindow, QApplication

class Ui_MainWindow(object):
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(400, 600)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        
        # Ajout d'un écran d'affichage pour les calculs
        self.display = QtWidgets.QLabel(self.centralwidget)
        self.display.setGeometry(QtCore.QRect(10, 10, 380, 50))
        self.display.setStyleSheet("background-color: rgb(255, 255, 255);")
        self.display.setAlignment(QtCore.Qt.AlignRight)
        self.display.setObjectName("display")
        self.display.setFont(QtGui.QFont("Arial", 14))
        
        # Configuration des boutons
        buttons = {
            '(': (10, 80), ')': (110, 80), 'C': (210, 80), '1/x': (310, 80),
            '7': (10, 140), '8': (110, 140), '9': (210, 140), '/': (310, 140),
            '4': (10, 200), '5': (110, 200), '6': (210, 200), 'X': (310, 200),
            '1': (10, 260), '2': (110, 260), '3': (210, 260), '-': (310, 260),
             '.': (110, 320), '0': (10, 320), '=': (210, 320), '+': (310, 320),
            '%': (10, 380), 'sin': (10, 440), 'cos': (110, 440), 'exp': (210, 440), 
            '!': (310, 440)
        }

        self.buttons = {}
        for btn_text, pos in buttons.items():
            self.buttons[btn_text] = QtWidgets.QPushButton(self.centralwidget)
            self.buttons[btn_text].setGeometry(QtCore.QRect(pos[0], pos[1], 90, 50))
            self.buttons[btn_text].setText(btn_text)
            self.buttons[btn_text].setFont(QtGui.QFont("Arial", 14))  # Augmenter la taille de la police
            self.buttons[btn_text].setStyleSheet("background-color: rgb(162, 255, 128);")
            self.buttons[btn_text].setObjectName(f"pushButton_{btn_text}")
            self.buttons[btn_text].clicked.connect(lambda ch, btn=btn_text: self.on_button_click(btn))

        # Ajoutez le bouton pour la racine carrée avec une icône
        self.buttons['sqrt'] = QtWidgets.QPushButton(self.centralwidget)
        self.buttons['sqrt'].setGeometry(QtCore.QRect(10, 380, 90, 50))
        self.buttons['sqrt'].setStyleSheet("background-color: rgb(162, 255, 128);")
        icon = QtGui.QIcon()
        icon.addPixmap(QtGui.QPixmap("C:\\Users\\elhad\\Desktop\\Calculatrice_Python\\icons\\racine-carree-de-x-signes-mathematiques.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.buttons['sqrt'].setIcon(icon)
        self.buttons['sqrt'].setObjectName("pushButton_sqrt")
        self.buttons['sqrt'].clicked.connect(self.calculate_sqrt)
        
        # Ajoutez le bouton pour "au carré" avec une icône
        self.buttons['^2'] = QtWidgets.QPushButton(self.centralwidget)
        self.buttons['^2'].setGeometry(QtCore.QRect(110, 380, 90, 50))
        self.buttons['^2'].setStyleSheet("background-color: rgb(162, 255, 128);")
        icon2 = QtGui.QIcon()
        icon2.addPixmap(QtGui.QPixmap("C:\\Users\\elhad\\Desktop\\Calculatrice_Python\\icons\\carré2.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.buttons['^2'].setIcon(icon2)
        self.buttons['^2'].setObjectName("pushButton_^2")
        self.buttons['^2'].clicked.connect(self.calculate_square)
        
        # Ajoutez le bouton pour "X puissance n" avec une icône
        self.buttons['^n'] = QtWidgets.QPushButton(self.centralwidget)
        self.buttons['^n'].setGeometry(QtCore.QRect(210, 380, 90, 50))
        self.buttons['^n'].setStyleSheet("background-color: rgb(162, 255, 128);")
        icon3 = QtGui.QIcon()
        icon3.addPixmap(QtGui.QPixmap("C:\\Users\\elhad\\Desktop\\Calculatrice_Python\\icons\\mathematiques.png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.buttons['^n'].setIcon(icon3)
        self.buttons['^n'].setObjectName("pushButton_^n")
        self.buttons['^n'].clicked.connect(self.on_power_click)
        
        # Ajoutez le bouton de retour en arrière avec une icône
        self.buttons['back'] = QtWidgets.QPushButton(self.centralwidget)
        self.buttons['back'].setGeometry(QtCore.QRect(310, 380, 90, 50))
        self.buttons['back'].setStyleSheet("background-color: rgb(162, 255, 128);")
        icon4 = QtGui.QIcon()
        icon4.addPixmap(QtGui.QPixmap("C:\\Users\\elhad\\Desktop\\Calculatrice_Python\\icons\\retour-arriere (1).png"), QtGui.QIcon.Normal, QtGui.QIcon.Off)
        self.buttons['back'].setIcon(icon4)
        self.buttons['back'].setObjectName("pushButton_back")
        self.buttons['back'].clicked.connect(self.backspace)

        MainWindow.setCentralWidget(self.centralwidget)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)
        self.menuBar = QtWidgets.QMenuBar(MainWindow)
        self.menuBar.setGeometry(QtCore.QRect(0, 0, 400, 26))
        self.menuBar.setObjectName("menuBar")
        self.menuCalculatrice = QtWidgets.QMenu(self.menuBar)
        self.menuCalculatrice.setObjectName("menuCalculatrice")
        MainWindow.setMenuBar(self.menuBar)
        self.menuBar.addAction(self.menuCalculatrice.menuAction())
        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "Calculatrice"))
        self.menuCalculatrice.setTitle(_translate("MainWindow", "Calculatrice"))

    def on_button_click(self, char):
        if char == "C":
            self.display.setText("")
        elif char == "=":
            self.evaluate_expression()
        elif char == "%":
            self.calculate_percentage()
        elif char == 'sqrt':
            self.calculate_sqrt()
        elif char == '^2':
            self.calculate_square()
        elif char == '^n':
            self.on_power_click()
        elif char == 'sin':
            self.calculate_sin()
        elif char == 'cos':
            self.calculate_cos()
        elif char == 'exp':
            self.calculate_exp()
        elif char == '!':
            self.calculate_factorial()
        elif char == 'back':
            self.backspace()
        else:
            current_text = self.display.text()
            new_text = current_text + char
            self.display.setText(new_text)
    
    def on_power_click(self):
        current_text = self.display.text()
        self.display.setText(current_text + '^')

    def evaluate_expression(self):
        try:
            expression = self.display.text().replace("X", "*").replace("^", "**").replace(",", ".")
            result = eval(expression)
            self.display.setText(str(result))
        except Exception as e:
            self.display.setText("Error")
    
    def calculate_sqrt(self):
        try:
            value = float(self.display.text())
            if value < 0:
                raise ValueError("Negative value")
            result = math.sqrt(value)
            self.display.setText(str(result))
        except ValueError:
            self.display.setText("Error")
        except Exception as e:
            self.display.setText("Error")

    def calculate_square(self):
        try:
            value = float(self.display.text())
            result = value ** 2
            self.display.setText(str(result))
        except Exception as e:
            self.display.setText("Error")
    
    def calculate_percentage(self):
        try:
            value = float(self.display.text())
            result = value / 100
            self.display.setText(str(result))
        except ValueError:
            self.display.setText("Error")
        except Exception as e:
            self.display.setText("Error")
    
    def calculate_sin(self):
        try:
            value = float(self.display.text())
            result = math.sin(math.radians(value))
            self.display.setText(str(result))
        except Exception as e:
            self.display.setText("Error")
    
    def calculate_cos(self):
        try:
            value = float(self.display.text())
            result = math.cos(math.radians(value))
            self.display.setText(str(result))
        except Exception as e:
            self.display.setText("Error")
    
    def calculate_exp(self):
        try:
            value = float(self.display.text())
            result = math.exp(value)
            self.display.setText(str(result))
        except Exception as e:
            self.display.setText("Error")
    
    def calculate_factorial(self):
        try:
            value = int(self.display.text())
            result = math.factorial(value)
            self.display.setText(str(result))
        except Exception as e:
            self.display.setText("Error")
    
    def backspace(self):
        current_text = self.display.text()
        self.display.setText(current_text[:-1])

    def calculate_power(self):
      try:
        expression = self.display.text()
        if '^' in expression:
            base, exponent = expression.split('^')
            base = float(base)
            exponent = float(exponent)
            result = base ** exponent
            self.display.setText(str(result))
        else:
            self.display.setText("Error")
      except Exception as e:
        self.display.setText("Error")

class Calculator(QMainWindow, Ui_MainWindow):
    def __init__(self):
        super().__init__()
        self.setupUi(self)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    MainWindow = Calculator()
    MainWindow.show()
    sys.exit(app.exec_())

```
